if-else is used to run code based on conditions.

```python
# Basic if-else example
age = 18

if age >= 18:
    print("You can vote")
else:
    print("You cannot vote")
```

```python
# if-elif-else example
marks = 85

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
else:
    print("Grade C")
```

```python
# Boolean condition example
is_sunday = True

if is_sunday:
    print("Have fun, today is holiday")
else:
    print("Go to work")
```

```python
# Check even or odd (use fixed value for quick testing)
num = 6

if num % 2 == 0:
    print("Number is even")
else:
    print("Number is odd")
```

```python

# Compare two numbers
a = 10
b = 20

if a > b:
    print("A is greater")
else:
    print("B is greater")
```

```python
# Check number type
num = -5

if num > 0:
    print("Positive number")
elif num < 0:
    print("Negative number")
else:
    print("Number is zero")
```

```python
# Calculator with safe division
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")

if op == "+":
    print("Result:", a + b)
elif op == "-":
    print("Result:", a - b)
elif op == "*":
    print("Result:", a * b)
elif op == "/":
    if b != 0:
        print("Result:", a / b)
    else:
        print("Cannot divide by zero")
else:
    print("Invalid operator")
```

```python
# Check pass or fail
marks = 45

if marks >= 50:
    print("Pass")
else:
    print("Fail")
```
