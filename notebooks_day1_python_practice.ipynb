1. Variables
Variables are used to store data in memory.

```python
x = "Ankit"
y = "boy"
z = 20

print("My name is", x)
print("I am a", y)
print("I am", z, "years old")
```

2. Data Types
Common data types in Python:
- int → whole numbers
- float → decimal numbers
- str → text
- bool → True or False 

```python
name = "Ankit"   # str
age = 20         # int
price = 99.5     # float
is_student = True  # bool

print(type(name))
print(type(age))
print(type(price))
print(type(is_student))
```

3. Arithmetic Operators
Basic math operations in Python.

```python
a = 2
b = 5

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Power:", a ** b)
print("Modulus:", a % b)
```

4. Comparison Operators
These are used to compare two values.

```python
a = 20
b = 6

print("a < b:", a < b)
print("a > b:", a > b)
print("a == b:", a == b)
print("a != b:", a != b)
print("a <= b:", a <= b)
print("a >= b:", a >= b)
```

5. Real-Life Example
Simple bill calculation.

```python
bottle_price = 20
extra_charge = 5

total = bottle_price + extra_charge

print("Final Bill is:", total)
```

6. Common Mistake and Fix
You cannot add a string and an integer directly.

```python
apples = 5
label = "Count: "

# Wrong way:
#print(label + apples)

# Correct way:
print(label + str(apples))
```

Common Mistakes & Fix 

Mistake 1: Adding string and integer
name = "apple"
number = 10

print(name + number)

Error: TypeError

 Reason:
Python cannot add different data types directly (string + integer)

 Fix:

print(name + str(number))

Output:apple10

---

Mistake 2: Wrong type assumption

x = "10"
y = 5

print(x + y)
Error: TypeError

 Reason:
"x" is a string, not a number

Fix:

print(int(x) + y)

Output:15

---

 Mistake 3: Division confusion

a = 5
b = 2

print(a / b)

Output: 2.5 (float)

Reason:
Normal division always returns float

Fix (if integer needed):

print(a // b)

Output:2

---

Mistake 4: Boolean confusion

print(True + True)

 Output: 2

 Reason:
True = 1 and False = 0 in Python

 Understanding:
True + True = 1 + 1 = 2
