String Indexing

Indexing is used to access characters from a string.
Index starts from 0.
Negative index starts from -1 for the last character.

```python
word = "PYTHON"

print(word[0])   # First character
print(word[3])   # Fourth character
print(word[-1])  # Last character
print(word[-2])  # Second last character
```

String Slicing

Slicing is used to get part of a string.

Format:
string[start:stop]

Note:
The stop index is not included.

```python
text = "Hello"

print(text[0:4])  # Hell
print(text[:3])   # Hel
print(text[3:])   # lo
print(text[:])    # Hello
```

String Methods

- upper()  → convert to capital letters
- lower()  → convert to small letters
- strip()   → remove extra spaces
- replace() → replace text

```python
msg = "   Hello Python   "

print(msg.upper())
print(msg.lower())
print(msg.strip())

sentence = "I like Apple"
print(sentence.replace("Apple", "Mango"))

# Method chaining
clean_text = msg.strip().upper()
print(clean_text)
```
